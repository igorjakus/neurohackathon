```python
import numpy as np
import pandas as pd
```

```python
ESR = pd.read_csv("../data/Epileptic_Seizure_Recognition.csv")
```

```python
def preprocess_data(dataset: pd.DataFrame) -> np.array:
    def split_id(id) -> tuple:
        # idk = i don't know what this value stands for
        try:
            time_stamp, idk, person = id.split(".")
        except Exception:
            time_stamp, idk = id.split(".")
            person = 999
        
        time_stamp = int(time_stamp[1:])
        person = int(person)
        return person, time_stamp, idk


    ids = dataset["Unnamed"]
    values = [tuple(vals) for vals in dataset.iloc[:, 1:179].to_numpy()]
    labels = dataset["y"]

    people, time_stamps = [], []
    for id in ids:
        person, time_stamp, _ = split_id(id)
        people.append(person)
        time_stamps.append(time_stamp)

    preprocessed = pd.DataFrame({
        "person" : people,
        "time_stamp" : time_stamps,
        "values" : values,
        "labels": labels,
    })

    preprocessed = preprocessed[preprocessed["person"] != 999]
    
    preprocessed_sorted = preprocessed.sort_values(by=["person", "time_stamp"])
    
    return preprocessed_sorted


preprocess_data(ESR)
```
