```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
ESR = pd.read_csv("../data/Epileptic_Seizure_Recognition.csv")
```

```python
array = ESR.iloc[:, [0, -1]].to_numpy()
```

```python
def split_id(id):
    try:
        timestamp, idk, person = id.split(".")
    except Exception:
        timestamp, idk = id.split(".")
        person = '999'
    
    timestamp = int(timestamp[1:])
    return person, timestamp, idk
```

```python
array
```

```python
def sort_by_person(array):
    return sorted(array, key=lambda row: split_id(row[0]))

gyat = sort_by_person(array)
gyat
```

```python
rows_with_1 = list(filter(lambda row: row[1] == 1, gyat))
rows_with_1
```

```python
seizure_people = {row[1] for row in rows_with_1}
seizure_people
```
